```python
# Sort plot in desired way
uniprot_ids = df['Uniprot ACC'].unique()
quantities = np.zeros([len(uniprot_ids)], np.float64) # stddevs[i] is the stddev of the ith uniprot_id
for (index, uniprot_id) in enumerate(uniprot_ids):
    nsamples = len(df.ix[df['Uniprot ACC']==uniprot_id, 'value'])
    mean = df.ix[df['Uniprot ACC']==uniprot_id, 'value'].mean()
    stddev = df.ix[df['Uniprot ACC']==uniprot_id, 'value'].std()
    quantities[index] = nsamples

indices = quantities.argsort()
sorted_uniprot_ids = [ uniprot_ids[index] for index in indices ]
```

```python
def AddBondsBetweenAtoms(rmol, abset):

    pred = oechem.OEIsAtomMember(abset.GetAtoms())
    for bond in rmol.GetBonds():
        if pred(bond.GetBgn()) and pred(bond.GetEnd()):
            abset.AddBond(bond)
```

```python


from openeye import oechem, oedepict
# Fluorescent scaffolds.
scaffold = 'c1cccc2c1cncn2' # quinazoline
substructure_search = oechem.OESubSearch(scaffold)
substructure_search.SetMaxMatches(1)
# Read ligand data.
filename = "../output/step10/chembl_bioact_ki_ic50_kd_summary_of_P00918.pkl"
ligand_dataframe = pd.read_pickle(filename)
# Convert numeric fields to numeric data.
ligand_dataframe = ligand_dataframe.convert_objects()
print ligand_dataframe.columns
ofs = oechem.oemolostream('quinazolines.mol2')
index = 0
for smiles in ligand_dataframe['smiles']:
    # Create molecule.
    molecule = oechem.OEGraphMol()
    oechem.OESmilesToMol(molecule, str(smiles))
    # Check for fluorescent scaffold.
    #oechem.OEPrepareSearch(molecule, substructure_search)
    matches = substructure_search.Match(molecule)
    matches = [ match for match in matches ]
    if len(matches) > 0:
#    if substructure_search.SingleMatch(molecule):
        print smiles
        oechem.OEWriteMolecule(ofs, molecule)
        # Show ligand.
        oedepict.OEPrepareDepiction(molecule)
        width, height = 300, 300
        opts = oedepict.OE2DMolDisplayOptions(width, height, oedepict.OEScale_AutoScale)
        disp = oedepict.OE2DMolDisplay(molecule, opts)
        # Get scaffold.            
        match = matches[0]
        highlight = oechem.OEAtomBondSet()
        for atom in match.GetAtoms():
            highlight.AddAtom(atom.target)
        AddBondsBetweenAtoms(molecule, highlight)
        # Highlight scaffold.
        oedepict.OEAddHighlighting(disp, oechem.OEYellowTint, oedepict.OEHighlightStyle_BallAndStick, highlight)
        # Render molecule.
        oedepict.OERenderMolecule("probe%05d.pdf" % index, disp)
        index += 1
        
ofs.close()
```
